## Phase 2: Complaint Data Acquisition & Preparation
Goal: Load civic complaint data as geographic points and prepare for spatial join


````python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import folium
import os
import warnings
warnings.filterwarnings('ignore')

# ── Reload Phase 1 data from disk (no re-downloading needed) ─────────────
edges = gpd.read_file("../data/roads_bandra.gpkg")
nodes = gpd.read_file("../data/nodes_bandra.gpkg")

print("Phase 1 data loaded successfully!")
print(f"  Road segments: {len(edges):,}")
print(f"  Nodes:         {len(nodes):,}")
print(f"  CRS:           {edges.crs}")
````

````python
np.random.seed(42)  

bbox = edges.total_bounds
print(f"Bandra West bounding box:")
print(f"  Longitude: {bbox[0]:.4f} to {bbox[2]:.4f}")
print(f"  Latitude:  {bbox[1]:.4f} to {bbox[3]:.4f}")

minx, miny, maxx, maxy = bbox
````

````python
# ── Layer 1: Baseline random complaints across the whole area ─────────────
n_baseline = 400

baseline_lons = np.random.uniform(minx, maxx, n_baseline)
baseline_lats = np.random.uniform(miny, maxy, n_baseline)

# ── Layer 2: Clustered complaints in high-density zones ───────────────────
# Real complaints cluster near markets, busy junctions, low-income areas
# We simulate this by adding Gaussian clusters at specific known hotspots

cluster_centres = [
    # (centre_lat, centre_lon, n_complaints, spread)
    # Near Bandra station area — very busy
    (19.0543, 72.8404, 80, 0.004),
    # Near Hill Road market
    (19.0612, 72.8347, 60, 0.003),
    # Near Linking Road
    (19.0659, 72.8378, 70, 0.003),
    # Near Turner Road
    (19.0589, 72.8301, 50, 0.002),
    # Near reclamation area — known waterlogging
    (19.0505, 72.8370, 90, 0.005),
]

cluster_lons = []
cluster_lats = []

for centre_lat, centre_lon, n, spread in cluster_centres:
    lons = np.random.normal(centre_lon, spread, n)
    lats = np.random.normal(centre_lat, spread, n)
    cluster_lons.extend(lons)
    cluster_lats.extend(lats)

# ── Combine baseline and clustered points ─────────────────────────────────
all_lons = np.concatenate([baseline_lons, cluster_lons])
all_lats = np.concatenate([baseline_lats, cluster_lats])
n_total  = len(all_lons)

print(f"Total complaint points generated: {n_total}")
print(f"  Baseline (random):  {n_baseline}")
print(f"  Clustered (hotspot): {len(cluster_lons)}")
````

````python
# ── Complaint types with realistic Mumbai proportions ─────────────────────
complaint_types = np.random.choice(
    ['Pothole', 'Damaged Road Surface', 'Waterlogging', 'Road Cave-in', 'Broken Footpath'],
    size=n_total,
    p=[0.40, 0.28, 0.18, 0.08, 0.06]   # probabilities must sum to 1.0
)

# ── Severity — most complaints are Medium ─────────────────────────────────
severity_levels = np.random.choice(
    ['Low', 'Medium', 'High'],
    size=n_total,
    p=[0.25, 0.55, 0.20]
)

# ── Month — with monsoon season spike ─────────────────────────────────────
# Months 6,7,8,9 (Jun–Sep) are monsoon — much higher complaint rates
# We weight these months to appear 3× more often
month_weights = [0.04, 0.04, 0.05, 0.06, 0.07,  # Jan–May
                 0.14, 0.18, 0.16, 0.12, 0.06,  # Jun–Oct (monsoon peak Jul–Sep)
                 0.05, 0.03]                      # Nov–Dec
month_weights = np.array(month_weights)
month_weights = month_weights / month_weights.sum()  # Normalise to sum to 1

months = np.random.choice(range(1, 13), size=n_total, p=month_weights)

# ── Derive monsoon flag from month ───────────────────────────────────────
is_monsoon = np.where(np.isin(months, [6, 7, 8, 9]), 1, 0)

# ── Ward names — Bandra West wards ───────────────────────────────────────
wards = np.random.choice(
    ['H/W-1', 'H/W-2', 'H/W-3', 'H/W-4'],
    size=n_total
)

# ── Build the DataFrame ───────────────────────────────────────────────────
complaints_df = pd.DataFrame({
    'complaint_id':   [f'BMC-SIM-{i:04d}' for i in range(n_total)],
    'longitude':      all_lons,
    'latitude':       all_lats,
    'complaint_type': complaint_types,
    'severity':       severity_levels,
    'month':          months,
    'is_monsoon':     is_monsoon,
    'ward':           wards,
})

print("Complaint DataFrame created!")
print(f"\nShape: {complaints_df.shape}")
print(f"\nComplaint type breakdown:")
print(complaints_df['complaint_type'].value_counts())
print(f"\nSeverity breakdown:")
print(complaints_df['severity'].value_counts())
print(f"\nMonsoon complaints: {is_monsoon.sum()} ({is_monsoon.mean()*100:.1f}%)")
````

````python
# ── CREATE complaints_gdf (MANDATORY FIRST STEP) ─────────────────────────

import pandas as pd
import geopandas as gpd

# Load your dataset (make sure file exists in /data folder)
df = complaints_df

# Normalize column names
df.columns = df.columns.str.lower()

# Fix column names if needed
if 'lat' in df.columns:
    df.rename(columns={'lat': 'latitude'}, inplace=True)
if 'lng' in df.columns:
    df.rename(columns={'lng': 'longitude'}, inplace=True)

# Check required columns
if 'latitude' not in df.columns or 'longitude' not in df.columns:
    raise ValueError("Dataset must contain latitude and longitude columns")

# Add complaint_id if missing
if 'complaint_id' not in df.columns:
    df['complaint_id'] = range(len(df))

# Drop missing coords
df = df.dropna(subset=['latitude', 'longitude'])

# Create GeoDataFrame
complaints_gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df['longitude'], df['latitude']),
    crs="EPSG:4326"
)

print(f"✅ complaints_gdf created with {len(complaints_gdf)} records")
````

````python
# ── Convert DataFrame to GeoDataFrame ─────────────────────────────────────

complaints_gdf = gpd.GeoDataFrame(
    complaints_df,
    geometry=gpd.points_from_xy(
        complaints_df['longitude'],   # X coordinate
        complaints_df['latitude']     # Y coordinate
    ),
    crs="EPSG:4326"   # WGS84 — same as your roads data
)

print("GeoDataFrame created!")
print(f"\nType:     {type(complaints_gdf)}")
print(f"Shape:    {complaints_gdf.shape}")
print(f"CRS:      {complaints_gdf.crs}")
print(f"\nGeometry column sample:")
print(complaints_gdf['geometry'].head(3))
````

````python
# ── Save complaint data ────────────────────────────────────────────────────

complaints_gdf.to_file("../data/complaints_bandra.gpkg", driver="GPKG")
complaints_df.to_csv("../data/complaints_bandra.csv", index=False)

print("Saved:")
print(f"  data/complaints_bandra.gpkg  ({len(complaints_gdf)} records)")
print(f"  data/complaints_bandra.csv   (for easy viewing)")
````

````python
# ── Static map: roads + complaints ────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# ── Left: All complaints coloured by type ─────────────────────────────────
ax1 = axes[0]
edges.plot(ax=ax1, color='#BDC3C7', linewidth=0.5, alpha=0.7, zorder=1)

type_colors = {
    'Pothole':               '#E74C3C',
    'Damaged Road Surface':  '#E67E22',
    'Waterlogging':          '#3498DB',
    'Road Cave-in':          '#8E44AD',
    'Broken Footpath':       '#27AE60',
}

for ctype, color in type_colors.items():
    subset = complaints_gdf[complaints_gdf['complaint_type'] == ctype]
    subset.plot(
        ax=ax1,
        color=color,
        markersize=4,
        alpha=0.7,
        zorder=2,
        label=f"{ctype} ({len(subset)})"
    )

ax1.set_title("Complaints by Type — Bandra West", fontsize=13, fontweight='bold')
ax1.set_axis_off()
ax1.legend(loc='lower left', fontsize=8, title="Complaint Type",
           framealpha=0.9, markerscale=1.5)

# ── Right: Complaints coloured by severity ────────────────────────────────
ax2 = axes[1]
edges.plot(ax=ax2, color='#BDC3C7', linewidth=0.5, alpha=0.7, zorder=1)

severity_colors = {
    'Low':    '#F9E79F',
    'Medium': '#F39C12',
    'High':   '#E74C3C',
}

for sev, color in severity_colors.items():
    subset = complaints_gdf[complaints_gdf['severity'] == sev]
    size   = {'Low': 3, 'Medium': 5, 'High': 8}[sev]
    subset.plot(
        ax=ax2,
        color=color,
        markersize=size,
        alpha=0.75,
        zorder=2,
        label=f"{sev} ({len(subset)})"
    )

ax2.set_title("Complaints by Severity — Bandra West", fontsize=13, fontweight='bold')
ax2.set_axis_off()
ax2.legend(loc='lower left', fontsize=9, title="Severity",
           framealpha=0.9, markerscale=1.5)

plt.suptitle("Phase 2 Complete — Complaint Data Overview",
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig("../outputs/phase2_complaints_map.png", dpi=150, bbox_inches='tight')
plt.show()

print("Map saved to outputs/phase2_complaints_map.png")
````

````python
# ── Exploratory summary ────────────────────────────────────────────────────

print("=" * 55)
print("  PHASE 2 SUMMARY — COMPLAINT DATA")
print("=" * 55)

print(f"\n  Total complaints:     {len(complaints_gdf):,}")
print(f"  Unique complaint IDs: {complaints_gdf['complaint_id'].nunique():,}")

print(f"\n  By type:")
for ctype, count in complaints_gdf['complaint_type'].value_counts().items():
    pct = count / len(complaints_gdf) * 100
    bar = '█' * int(pct / 2.5)
    print(f"    {ctype:<25} {count:>4}  {bar} {pct:.1f}%")

print(f"\n  By severity:")
for sev, count in complaints_gdf['severity'].value_counts().items():
    pct = count / len(complaints_gdf) * 100
    print(f"    {sev:<10} {count:>4}  ({pct:.1f}%)")

monsoon = complaints_gdf['is_monsoon'].sum()
print(f"\n  Monsoon complaints (Jun–Sep): {monsoon} ({monsoon/len(complaints_gdf)*100:.1f}%)")

peak_month_num = complaints_gdf['month'].value_counts().index[0]
month_names    = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
                  7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
print(f"  Peak complaint month:         {month_names[peak_month_num]}")

print(f"\n  Geographic spread:")
print(f"    Lat range:  {complaints_gdf['latitude'].min():.4f} – {complaints_gdf['latitude'].max():.4f}")
print(f"    Lon range:  {complaints_gdf['longitude'].min():.4f} – {complaints_gdf['longitude'].max():.4f}")

print(f"\n  Files saved:")
print(f"    data/complaints_bandra.gpkg")
print(f"    data/complaints_bandra.csv")
print(f"    outputs/phase2_complaints_map.png")
print(f"\n  Next: Phase 3 — Spatial join: snap complaints to road segments")
````

## Phase 3: Spatial Join — Connecting Complaints to Road Segments
Goal: Snap each complaint point to its nearest road segment and count complaints per road


````python
# ── Phase 3 Imports ───────────────────────────────────────────────────────
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings
warnings.filterwarnings('ignore')

# ── Reload saved data from Phase 1 and Phase 2 ───────────────────────────
edges         = gpd.read_file("../data/roads_bandra.gpkg")
nodes         = gpd.read_file("../data/nodes_bandra.gpkg")
complaints_gdf = gpd.read_file("../data/complaints_bandra.gpkg")

print("All Phase 1 & 2 data loaded!")
print(f"  Road segments:  {len(edges):,}")
print(f"  Complaint pts:  {len(complaints_gdf):,}")
print(f"  Roads CRS:      {edges.crs}")
print(f"  Complaints CRS: {complaints_gdf.crs}")
````

````python
# ── Reproject to UTM Zone 43N (metres) for accurate distance calculation ──

edges_m      = edges.to_crs("EPSG:32643")
complaints_m = complaints_gdf.to_crs("EPSG:32643")

print("Reprojection complete!")
print(f"  Roads CRS:      {edges_m.crs}")
print(f"  Complaints CRS: {complaints_m.crs}")

# Prove that coordinates changed
print(f"\n  Sample road coord (EPSG:4326):  "
      f"{edges.geometry.iloc[0].coords[0]}")
print(f"  Same road coord (EPSG:32643):  "
      f"{edges_m.geometry.iloc[0].coords[0]}")
````

````python
# ── Prepare edges for spatial join ────────────────────────────────────────

# Reset the multi-level index to get osmid as a regular column
edges_m_reset = edges_m.reset_index()

# Check what columns exist
print("All edge columns:")
print(edges_m_reset.columns.tolist())

# Keep only what you need — avoid bloating the join result
keep_cols = ['osmid', 'name', 'highway', 'length', 'geometry']

# Only keep columns that actually exist in your data
keep_cols = [c for c in keep_cols if c in edges_m_reset.columns]

edges_for_join = edges_m_reset[keep_cols].copy()

print(f"\nEdges prepared for join:")
print(f"  Rows:    {len(edges_for_join):,}")
print(f"  Columns: {edges_for_join.columns.tolist()}")
print(f"  CRS:     {edges_for_join.crs}")

# Check for any null geometries in edges (these cause join errors)
null_edges = edges_for_join['geometry'].isna().sum()
print(f"  Null geometries: {null_edges} "
      f"({'OK' if null_edges == 0 else 'PROBLEM — remove these'})")

if null_edges > 0:
    edges_for_join = edges_for_join.dropna(subset=['geometry'])
    print(f"  Null geometries removed. Remaining: {len(edges_for_join):,}")
````

````python
# ── SPATIAL JOIN: snap each complaint to its nearest road segment ──────────

print("Running spatial join...")
print("For each complaint point, finding the nearest road segment within 50m...")
print()

complaints_joined = gpd.sjoin_nearest(
    complaints_m,           # LEFT:  the points (complaints)
    edges_for_join,         # RIGHT: the lines (roads)
    how='left',             # Keep ALL complaints even if no road found nearby
    max_distance=50,        # Only match if road is within 50 metres
    distance_col='dist_to_road'  # Store the actual distance in a new column
)

print("Spatial join complete!")
print(f"\nResult shape: {complaints_joined.shape}")
print(f"Columns added from roads: {[c for c in complaints_joined.columns if c not in complaints_m.columns]}")
````

````python
# ── Join quality check ────────────────────────────────────────────────────

total         = len(complaints_joined)
matched       = complaints_joined['osmid'].notna().sum()
unmatched     = complaints_joined['osmid'].isna().sum()
match_rate    = matched / total * 100

print("=== SPATIAL JOIN QUALITY REPORT ===\n")
print(f"  Total complaints:          {total:,}")
print(f"  Matched to a road:         {matched:,}  ({match_rate:.1f}%)")
print(f"  Unmatched (>50m from road):{unmatched:,}  ({100-match_rate:.1f}%)")

if unmatched > 0:
    print(f"\n  Note: {unmatched} complaints were more than 50m from any road.")
    print(f"  These are likely in parks, water bodies, or at the area boundary.")
    print(f"  They will be dropped — this is expected and acceptable.")

print(f"\n  Distance to matched road (metres):")
dist_stats = complaints_joined['dist_to_road'].dropna().describe()
print(f"    Mean:   {dist_stats['mean']:.1f}m")
print(f"    Median: {dist_stats['50%']:.1f}m")
print(f"    Max:    {dist_stats['max']:.1f}m")
````

````python
# ── Remove unmatched complaints ────────────────────────────────────────────

complaints_clean = complaints_joined.dropna(subset=['osmid']).copy()

# 🔥 FIX: safely convert osmid to integer
# Sometimes osmid can be float or list → handle properly

def clean_osmid(x):
    if isinstance(x, list):
        return int(x[0])  # take first if multiple
    try:
        return int(x)
    except:
        return None

complaints_clean['osmid'] = complaints_clean['osmid'].apply(clean_osmid)

# Drop any still-invalid osmids
complaints_clean = complaints_clean.dropna(subset=['osmid'])

# Remove duplicate matches — keep closest road
complaints_clean = complaints_clean.sort_values('dist_to_road')
complaints_clean = complaints_clean.drop_duplicates(
    subset=['complaint_id'], keep='first'
)

print(f"Clean join result: {len(complaints_clean):,} complaints matched to roads")

print("\nSample — first 5 rows showing complaint + matched road:")
display_cols = ['complaint_id', 'complaint_type', 'severity',
                'name', 'highway', 'dist_to_road']
display_cols = [c for c in display_cols if c in complaints_clean.columns]

print(complaints_clean[display_cols].head())
````

````python
# ── Aggregate complaints by road segment ──────────────────────────────────

complaint_counts = complaints_clean.groupby('osmid').agg(

    total_complaints    = ('complaint_type',  'count'),

    pothole_count       = ('complaint_type',
                           lambda x: (x == 'Pothole').sum()),

    damaged_road_count  = ('complaint_type',
                           lambda x: (x == 'Damaged Road Surface').sum()),

    waterlog_count      = ('complaint_type',
                           lambda x: (x == 'Waterlogging').sum()),

    high_severity_count = ('severity',
                           lambda x: (x == 'High').sum()),

    monsoon_complaints  = ('is_monsoon',      'sum'),

    avg_dist_to_road    = ('dist_to_road',    'mean'),

).reset_index()

print("Complaint counts per road segment:")
print(f"  Road segments with at least 1 complaint: {len(complaint_counts):,}")
print(f"  Total complaints accounted for:          "
      f"{complaint_counts['total_complaints'].sum():,}")

print(f"\nDistribution of complaints per road:")
print(complaint_counts['total_complaints'].describe().round(2))

print(f"\nTop 5 most complained-about roads:")
top5 = complaint_counts.nlargest(5, 'total_complaints')
print(top5[['osmid', 'total_complaints', 'pothole_count', 'high_severity_count']])
````

````python
# ── FIX OSMID TYPES BEFORE MERGE ─────────────────────────────────────────

# Ensure edges osmid is clean
def clean_osmid(x):
    if isinstance(x, list):
        return int(x[0])
    try:
        return int(x)
    except:
        return None

edges_m_reset['osmid'] = edges_m_reset['osmid'].apply(clean_osmid)
complaint_counts['osmid'] = complaint_counts['osmid'].apply(clean_osmid)

# Drop invalid osmids
edges_m_reset = edges_m_reset.dropna(subset=['osmid'])
complaint_counts = complaint_counts.dropna(subset=['osmid'])

# Convert to int (safe now)
edges_m_reset['osmid'] = edges_m_reset['osmid'].astype(int)
complaint_counts['osmid'] = complaint_counts['osmid'].astype(int)

# ── MERGE ────────────────────────────────────────────────────────────────

roads_with_counts = edges_m_reset.merge(
    complaint_counts,
    on='osmid',
    how='left'
)

# ── FILL MISSING VALUES ──────────────────────────────────────────────────

fill_cols = ['total_complaints', 'pothole_count', 'damaged_road_count',
             'waterlog_count', 'high_severity_count', 'monsoon_complaints']

for col in fill_cols:
    if col in roads_with_counts.columns:
        roads_with_counts[col] = roads_with_counts[col].fillna(0).astype(int)

# ── DEBUG INFO ───────────────────────────────────────────────────────────

print("✅ Merge complete!")
print(f"Total road segments: {len(roads_with_counts):,}")

if 'total_complaints' in roads_with_counts.columns:
    print(f"Roads with complaints: {(roads_with_counts['total_complaints'] > 0).sum():,}")
    print(f"Roads with zero complaints: {(roads_with_counts['total_complaints'] == 0).sum():,}")
    print(f"Total complaints accounted: {roads_with_counts['total_complaints'].sum():,}")

print("\nSample rows:")
display_cols = ['name', 'highway', 'length',
                'total_complaints', 'pothole_count',
                'high_severity_count']
display_cols = [c for c in display_cols if c in roads_with_counts.columns]

print(roads_with_counts[display_cols].head(8))
````

````python
# ── Validate the merge result ──────────────────────────────────────────────

print("=== MERGE VALIDATION ===\n")

# Check 1: No rows lost
print(f"[{'PASS' if len(roads_with_counts) == len(edges_m_reset) else 'FAIL'}] "
      f"Row count preserved: {len(roads_with_counts):,} roads")

# Check 2: Complaint totals add up
total_in  = len(complaints_clean)
total_out = roads_with_counts['total_complaints'].sum()
print(f"[{'PASS' if total_in == total_out else 'FAIL'}] "
      f"Complaint total matches: {total_out:,} (expected {total_in:,})")

# Check 3: No negative values
neg_counts = (roads_with_counts['total_complaints'] < 0).sum()
print(f"[{'PASS' if neg_counts == 0 else 'FAIL'}] "
      f"No negative complaint counts: {neg_counts}")

# Check 4: Geometry still intact
null_geom = roads_with_counts['geometry'].isna().sum()
print(f"[{'PASS' if null_geom == 0 else 'FAIL'}] "
      f"Geometry intact: {null_geom} null geometries")

# Check 5: CRS still correct
print(f"[INFO] CRS: {roads_with_counts.crs} "
      f"({'correct' if roads_with_counts.crs.to_epsg() == 32643 else 'check this'})")
````

````python
# ── Convert back to EPSG:4326 for mapping ────────────────────────────────
roads_with_counts_4326 = roads_with_counts.to_crs("EPSG:4326")
complaints_clean_4326  = complaints_clean.to_crs("EPSG:4326")

# ── Plot ──────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# ── Left: complaint count heatmap on roads ────────────────────────────────
ax1 = axes[0]

# Roads with no complaints — grey
no_complaint = roads_with_counts_4326[
    roads_with_counts_4326['total_complaints'] == 0
]
no_complaint.plot(ax=ax1, color='#EAECEE', linewidth=0.5, zorder=1)

# Roads with complaints — colour by count
has_complaint = roads_with_counts_4326[
    roads_with_counts_4326['total_complaints'] > 0
]
has_complaint.plot(
    ax=ax1,
    column='total_complaints',
    cmap='YlOrRd',          # Yellow → Orange → Red
    linewidth=1.5,
    legend=True,
    legend_kwds={'label': 'Complaint Count', 'shrink': 0.6},
    zorder=2
)

# Complaint points on top
complaints_clean_4326.plot(
    ax=ax1,
    color='#2C3E50',
    markersize=2,
    alpha=0.4,
    zorder=3
)

ax1.set_title("Complaints per Road Segment\n(raw count)", fontsize=12, fontweight='bold')
ax1.set_axis_off()

# ── Right: pothole count specifically ─────────────────────────────────────
ax2 = axes[1]

no_complaint.plot(ax=ax2, color='#EAECEE', linewidth=0.5, zorder=1)

pothole_roads = roads_with_counts_4326[
    roads_with_counts_4326['pothole_count'] > 0
]
pothole_roads.plot(
    ax=ax2,
    column='pothole_count',
    cmap='Reds',
    linewidth=1.5,
    legend=True,
    legend_kwds={'label': 'Pothole Count', 'shrink': 0.6},
    zorder=2
)

ax2.set_title("Pothole Reports per Road Segment", fontsize=12, fontweight='bold')
ax2.set_axis_off()

plt.suptitle("Phase 3 Complete — Complaints Joined to Roads",
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig("../outputs/phase3_complaints_on_roads.png", dpi=150, bbox_inches='tight')
plt.show()
print("Map saved!")
````

## Phase 4: Risk Score Engineering
Goal: Transform raw complaint counts into a weighted, normalised 0–10 risk score

````python
# ── Phase 4 Imports ───────────────────────────────────────────────────────
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

# ── Reload Phase 3 output ─────────────────────────────────────────────────
roads = gpd.read_file("../data/roads_with_counts.gpkg")

print("Phase 3 data loaded!")
print(f"  Road segments:  {len(roads):,}")
print(f"  CRS:            {roads.crs}")
print(f"\nColumns available:")
print(roads.columns.tolist())
print(f"\nComplaint column check:")
print(roads[['total_complaints','pothole_count',
             'high_severity_count','monsoon_complaints']].describe().round(2))
````

````python
# ── Factor 1: Pothole density ──────────────────────────────────────────────

# Length is in metres — convert to km first
roads['length_km'] = roads['length'] / 1000

# Complaints per km — clip lower bound to avoid division by zero
# A road shorter than 10m gets treated as 10m long
roads['pothole_density'] = (
    roads['pothole_count'] /
    roads['length_km'].clip(lower=0.01)
)

print("Factor 1: Pothole Density")
print(f"  Min:    {roads['pothole_density'].min():.2f} complaints/km")
print(f"  Max:    {roads['pothole_density'].max():.2f} complaints/km")
print(f"  Mean:   {roads['pothole_density'].mean():.2f} complaints/km")
print(f"  Median: {roads['pothole_density'].median():.2f} complaints/km")

# Roads with non-zero pothole density
nonzero = (roads['pothole_density'] > 0).sum()
print(f"\n  Roads with at least 1 pothole complaint: {nonzero:,} "
      f"({nonzero/len(roads)*100:.1f}%)")

# Visualise the distribution
fig, ax = plt.subplots(figsize=(10, 4))
roads[roads['pothole_density'] > 0]['pothole_density'].hist(
    bins=30, ax=ax, color='#E74C3C', edgecolor='white', alpha=0.85
)
ax.set_xlabel('Pothole Complaints per km')
ax.set_ylabel('Number of Road Segments')
ax.set_title('Distribution of Pothole Density Across Road Segments')
ax.axvline(roads['pothole_density'].mean(), color='black',
           linestyle='--', linewidth=1.5, label='Mean')
ax.legend()
plt.tight_layout()
plt.savefig("../outputs/phase4_factor1_distribution.png", dpi=150, bbox_inches='tight')
plt.show()

print("\nHistogram saved!")
````

````python
# ── Factor 2: Traffic volume proxy from road type ─────────────────────────

# These scores are based on typical urban traffic volumes by road class
# Primary roads carry ~8x more traffic than residential lanes
traffic_proxy_map = {
    'motorway':      10,
    'motorway_link':  9,
    'trunk':          9,
    'trunk_link':     8,
    'primary':        8,
    'primary_link':   7,
    'secondary':      6,
    'secondary_link': 5,
    'tertiary':       4,
    'tertiary_link':  3,
    'residential':    2,
    'living_street':  1,
    'unclassified':   2,
    'service':        1,
    'pedestrian':     0,
    'track':          0,
}

def get_traffic_proxy(highway_val):
    # highway column can be a string or a list — handle both
    if isinstance(highway_val, list):
        highway_val = highway_val[0]
    highway_val = str(highway_val).strip().lower()
    return traffic_proxy_map.get(highway_val, 3)  # default 3 if unknown

roads['traffic_proxy'] = roads['highway'].apply(get_traffic_proxy)

print("Factor 2: Traffic Volume Proxy")
print(f"\nProxy score by road type:")
type_summary = (roads.groupby('highway')['traffic_proxy']
                     .first()
                     .sort_values(ascending=False))
for highway, score in type_summary.items():
    count = (roads['highway'] == highway).sum()
    print(f"  {str(highway):<20} score={score:>2}   ({count} segments)")
````

````python
# ── Factor 3: Road vulnerability to degradation ───────────────────────────

vulnerability_map = {
    'motorway':       1,   # Built to highest spec, best maintained
    'motorway_link':  2,
    'trunk':          2,
    'trunk_link':     3,
    'primary':        3,
    'primary_link':   4,
    'secondary':      5,
    'secondary_link': 6,
    'tertiary':       7,
    'tertiary_link':  7,
    'residential':   10,   # Lowest spec, least maintained, highest vulnerability
    'living_street':  9,
    'unclassified':   8,
    'service':        7,
    'pedestrian':     6,
    'track':          5,
}

def get_vulnerability(highway_val):
    if isinstance(highway_val, list):
        highway_val = highway_val[0]
    highway_val = str(highway_val).strip().lower()
    return vulnerability_map.get(highway_val, 5)

roads['road_vulnerability'] = roads['highway'].apply(get_vulnerability)

print("Factor 3: Road Vulnerability")
print(f"\nVulnerability score by road type:")
vuln_summary = (roads.groupby('highway')['road_vulnerability']
                     .first()
                     .sort_values(ascending=False))
for highway, score in vuln_summary.items():
    count = (roads['highway'] == highway).sum()
    bar   = '█' * score
    print(f"  {str(highway):<20} {bar:<12} {score:>2}   ({count} segments)")
````

````python
# ── Normalise all factors to 0–1 scale ────────────────────────────────────

scaler = MinMaxScaler(feature_range=(0, 1))

# Normalise Factor 1
roads['f1_norm'] = scaler.fit_transform(
    roads[['pothole_density']]
)

# Normalise Factor 2
roads['f2_norm'] = scaler.fit_transform(
    roads[['traffic_proxy']]
)

# Normalise Factor 3
roads['f3_norm'] = scaler.fit_transform(
    roads[['road_vulnerability']]
)

print("All factors normalised to 0–1 scale")
print(f"\nNormalised factor ranges:")
print(f"  F1 pothole density:    {roads['f1_norm'].min():.3f} – {roads['f1_norm'].max():.3f}")
print(f"  F2 traffic proxy:      {roads['f2_norm'].min():.3f} – {roads['f2_norm'].max():.3f}")
print(f"  F3 road vulnerability: {roads['f3_norm'].min():.3f} – {roads['f3_norm'].max():.3f}")

# Visualise all three normalised factors side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, col, label, color in zip(
    axes,
    ['f1_norm', 'f2_norm', 'f3_norm'],
    ['F1: Pothole Density (50%)',
     'F2: Traffic Volume (30%)',
     'F3: Road Vulnerability (20%)'],
    ['#E74C3C', '#3498DB', '#F39C12']
):
    roads[col].hist(bins=20, ax=ax, color=color, edgecolor='white', alpha=0.85)
    ax.set_title(label, fontsize=11, fontweight='bold')
    ax.set_xlabel('Normalised Score (0–1)')
    ax.set_ylabel('Road Segments')
    ax.axvline(roads[col].mean(), color='black',
               linestyle='--', linewidth=1.2, label=f"Mean: {roads[col].mean():.2f}")
    ax.legend(fontsize=9)

plt.suptitle('Distribution of Normalised Risk Factors',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layo
````

````python
# ── THE RISK FORMULA ──────────────────────────────────────────────────────
#
#   Raw Risk = 0.5 × F1 (pothole density)
#            + 0.3 × F2 (traffic volume)
#            + 0.2 × F3 (road vulnerability)
#

W1, W2, W3 = 0.5, 0.3, 0.2   # Weights — must sum to 1.0
assert W1 + W2 + W3 == 1.0, "Weights must sum to 1.0"

roads['raw_risk'] = (
    W1 * roads['f1_norm'] +
    W2 * roads['f2_norm'] +
    W3 * roads['f3_norm']
)

print("Weighted formula applied!")
print(f"  Formula: {W1} × F1 + {W2} × F2 + {W3} × F3")
print(f"\n  Raw risk score range: {roads['raw_risk'].min():.4f} – {roads['raw_risk'].max():.4f}")
print(f"  Raw risk score mean:  {roads['raw_risk'].mean():.4f}")
````

````python
# ── Monsoon multiplier ────────────────────────────────────────────────────

# Roads above the 75th percentile of monsoon complaints get a 20% boost
monsoon_threshold = roads['monsoon_complaints'].quantile(0.75)

print(f"Monsoon complaint threshold (75th percentile): {monsoon_threshold:.0f} complaints")

roads['monsoon_multiplier'] = roads['monsoon_complaints'].apply(
    lambda x: 1.20 if x > monsoon_threshold else 1.0
)

roads_with_boost = (roads['monsoon_multiplier'] > 1.0).sum()
print(f"Roads receiving monsoon boost: {roads_with_boost:,} "
      f"({roads_with_boost/len(roads)*100:.1f}%)")

# Apply monsoon multiplier to raw risk
roads['raw_risk_boosted'] = roads['raw_risk'] * roads['monsoon_multiplier']

print(f"\nRaw risk after monsoon boost:")
print(f"  Min:  {roads['raw_risk_boosted'].min():.4f}")
print(f"  Max:  {roads['raw_risk_boosted'].max():.4f}")
````

````python
# ── Final normalisation to 0–10 scale ─────────────────────────────────────

final_scaler = MinMaxScaler(feature_range=(0, 10))

roads['risk_score'] = final_scaler.fit_transform(
    roads[['raw_risk_boosted']]
).round(2)

print("Final risk score computed!")
print(f"\nRisk score statistics:")
print(roads['risk_score'].describe().round(2))
````

````python
# ── Risk categories ────────────────────────────────────────────────────────

def categorise_risk(score):
    if score >= 7.0:   return 'High'
    elif score >= 4.0: return 'Medium'
    elif score > 0.0:  return 'Low'
    else:              return 'No Data'

roads['risk_category'] = roads['risk_score'].apply(categorise_risk)

# Summary
cat_counts = roads['risk_category'].value_counts()
print("Risk Category Breakdown:")
print()
for cat, count in cat_counts.items():
    pct = count / len(roads) * 100
    bar = '█' * int(pct / 2)
    color_label = {'High':'🔴', 'Medium':'🟡', 'Low':'🟢', 'No Data':'⚪'}
    print(f"  {color_label.get(cat,'')} {cat:<10} {count:>4} segments   "
          f"{bar} {pct:.1f}%")
````

````python
# ── Factor contribution analysis ──────────────────────────────────────────

print("=== FACTOR CONTRIBUTION BY RISK CATEGORY ===\n")

factor_analysis = roads.groupby('risk_category').agg(
    count              = ('risk_score',       'count'),
    avg_risk_score     = ('risk_score',       'mean'),
    avg_f1_pothole     = ('f1_norm',          'mean'),
    avg_f2_traffic     = ('f2_norm',          'mean'),
    avg_f3_vulnerability=('f3_norm',          'mean'),
    avg_total_complaints=('total_complaints', 'mean'),
    avg_pothole_count  = ('pothole_count',    'mean'),
    monsoon_boost_pct  = ('monsoon_multiplier',
                          lambda x: (x > 1.0).mean() * 100)
).round(3)

print(factor_analysis.to_string())

# Visualise factor contributions as stacked bar
fig, ax = plt.subplots(figsize=(10, 5))

categories  = ['Low', 'Medium', 'High']
f1_vals     = [factor_analysis.loc[c, 'avg_f1_pothole']     for c in categories
               if c in factor_analysis.index]
f2_vals     = [factor_analysis.loc[c, 'avg_f2_traffic']     for c in categories
               if c in factor_analysis.index]
f3_vals     = [factor_analysis.loc[c, 'avg_f3_vulnerability']for c in categories
               if c in factor_analysis.index]
cats_avail  = [c for c in categories if c in factor_analysis.index]

x    = np.arange(len(cats_avail))
w    = 0.25

ax.bar(x - w,   f1_vals, w, label='F1: Pothole Density (50%)',    color='#E74C3C', alpha=0.85)
ax.bar(x,       f2_vals, w, label='F2: Traffic Volume (30%)',      color='#3498DB', alpha=0.85)
ax.bar(x + w,   f3_vals, w, label='F3: Road Vulnerability (20%)', color='#F39C12', alpha=0.85)

ax.set_xticks(x)
ax.set_xticklabels(cats_avail, fontsize=12)
ax.set_ylabel('Average Normalised Factor Value (0–1)')
ax.set_title('What Drives Each Risk Category', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.set_ylim(0, 1.1)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig("../outputs/phase4_factor_contributions.png", dpi=150, bbox_inches='tight')
plt.show()
print("Factor contribution chart saved!")
````

````python
# ── Risk score map ─────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

def risk_color(score):
    if score >= 7.0:  return '#E74C3C'   # Red
    elif score >= 4.0:return '#F39C12'   # Orange
    elif score > 0.0: return '#27AE60'   # Green
    else:             return '#BDC3C7'   # Grey

roads['plot_color'] = roads['risk_score'].apply(risk_color)

# ── Left: Risk category map ───────────────────────────────────────────────
ax1 = axes[0]
for color, label in [('#E74C3C','High Risk'),
                      ('#F39C12','Medium Risk'),
                      ('#27AE60','Low Risk'),
                      ('#BDC3C7','No Data')]:
    subset = roads[roads['plot_color'] == color]
    if len(subset) > 0:
        lw = 2.5 if color == '#E74C3C' else 1.5
        subset.plot(ax=ax1, color=color, linewidth=lw,
                    alpha=0.85, label=f"{label} ({len(subset)})")

ax1.set_title("Road Risk Categories", fontsize=13, fontweight='bold')
ax1.legend(loc='lower left', fontsize=9, framealpha=0.9)
ax1.set_axis_off()

# ── Right: Continuous risk score choropleth ───────────────────────────────
ax2 = axes[1]
roads[roads['risk_score'] == 0].plot(
    ax=ax2, color='#BDC3C7', linewidth=0.5, alpha=0.7
)
roads[roads['risk_score'] > 0].plot(
    ax=ax2,
    column='risk_score',
    cmap='RdYlGn_r',    # Red=high risk, Green=low risk
    linewidth=1.5,
    legend=True,
    vmin=0, vmax=10,
    legend_kwds={
        'label': 'Risk Score (0–10)',
        'shrink': 0.6,
        'orientation': 'vertical'
    }
)
ax2.set_title("Risk Score (0–10 continuous)", fontsize=13, fontweight='bold')
ax2.set_axis_off()

plt.suptitle("Phase 4 Complete — Road Risk Scores",
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig("../outputs/phase4_risk_score_map.png", dpi=150, bbox_inches='tight')
plt.show()
print("Risk score map saved!")
````

````python
# ── Save complete scored roads table ──────────────────────────────────────

roads.to_file("../data/roads_scored.gpkg", driver="GPKG")

roads.drop(columns=['geometry']).to_csv(
    "../data/roads_scored.csv", index=False
)

print("Phase 4 outputs saved!")
print("  data/roads_scored.gpkg  — full scored GeoDataFrame")
print("  data/roads_scored.csv   — for viewing in Excel")
print()
print("=== PHASE 4 SUMMARY ===")
print(f"  Formula:  {W1}×(pothole density) + "
      f"{W2}×(traffic volume) + {W3}×(road vulnerability)")
print(f"  Monsoon multiplier: ×1.2 on top 25% monsoon roads")
print(f"  Score range: 0–10  (higher = more dangerous)")
print()
cat_counts = roads['risk_category'].value_counts()
for cat in ['High','Medium','Low','No Data']:
    if cat in cat_counts.index:
        n   = cat_counts[cat]
        pct = n / len(roads) * 100
        print(f"  {cat:<10} {n:>4} roads  ({pct:.1f}%)")
print()
top = roads.nlargest(3, 'risk_score')[['name','highway','risk_score','risk_category']]
print("  Top 3 highest risk roads:")
print(top.to_string(index=False))
print()
print("  Next: Phase 5 — Build the interactive Folium map")
````

## Phase 5: Interactive Map — Folium Dashboard
Goal: Build a professional interactive web map of the road risk analysis

````python
# ── Phase 5 Imports ───────────────────────────────────────────────────────
import geopandas as gpd
import pandas as pd
import numpy as np
import folium
from folium.plugins import MiniMap, Fullscreen, LocateControl
import warnings
import os
warnings.filterwarnings('ignore')

# ── Reload scored roads from Phase 4 ─────────────────────────────────────
roads = gpd.read_file("../data/roads_scored.gpkg")
complaints = gpd.read_file("../data/complaints_bandra.gpkg")

# Make sure both are in EPSG:4326 for web mapping
roads      = roads.to_crs("EPSG:4326")
complaints = complaints.to_crs("EPSG:4326")

print("Data loaded and confirmed in EPSG:4326!")
print(f"  Road segments:  {len(roads):,}")
print(f"  Complaints:     {len(complaints):,}")
print(f"  Roads CRS:      {roads.crs}")
print(f"\nRisk category breakdown:")
print(roads['risk_category'].value_counts())
````

````python
# ── Colour mapping functions ───────────────────────────────────────────────

def get_road_color(risk_score):
    """Return hex colour based on risk score 0–10"""
    if pd.isna(risk_score) or risk_score == 0:
        return '#BDC3C7'   # Grey — no data
    elif risk_score >= 7.0:
        return '#E74C3C'   # Red — High risk
    elif risk_score >= 4.0:
        return '#F39C12'   # Orange — Medium risk
    else:
        return '#27AE60'   # Green — Low risk

def get_road_weight(highway_val):
    """Return line thickness based on road importance"""
    if isinstance(highway_val, list):
        highway_val = highway_val[0]
    highway_val = str(highway_val).lower()
    weight_map = {
        'motorway':   6,
        'trunk':      5,
        'primary':    4,
        'secondary':  3,
        'tertiary':   2.5,
        'residential':2,
        'unclassified':1.5,
        'service':    1.5,
    }
    return weight_map.get(highway_val, 2)

def get_road_opacity(risk_score):
    """Higher risk roads are more opaque — they stand out more"""
    if pd.isna(risk_score) or risk_score == 0:
        return 0.35
    elif risk_score >= 7.0:
        return 0.95
````

````python
# ── Base map ───────────────────────────────────────────────────────────────

# Calculate the centre of your road network
centre_lat = roads.geometry.centroid.y.mean()
centre_lon = roads.geometry.centroid.x.mean()

print(f"Map centre: {centre_lat:.4f}, {centre_lon:.4f}")

m = folium.Map(
    location=[centre_lat, centre_lon],
    zoom_start=14,
    tiles='CartoDB positron',
    prefer_canvas=True   # Renders using HTML Canvas instead of SVG
                         # Much faster for 1000+ line features
)

# ── Add map controls ───────────────────────────────────────────────────────
Fullscreen(
    position='topright',
    title='Full screen',
    title_cancel='Exit full screen'
).add_to(m)

MiniMap(
    tile_layer='CartoDB positron',
    position='bottomright',
    width=150,
    height=150,
    zoom_level_offset=-5
).add_to(m)

print("Base map created!")
print(f"  Centre: [{centre_lat:.4f}, {centre_lon:.4f}]")
print(f"  Zoom:   14")
print(f"  Tiles:  CartoDB positron")
````

````python
# ── Layer 1: Road risk segments ────────────────────────────────────────────

road_layer = folium.FeatureGroup(
    name="🛣️ Road Risk Score",
    show=True   # Visible by default when map opens
)

skipped = 0
added   = 0

for idx, row in roads.iterrows():

    # Skip rows with no geometry
    if row.geometry is None or row.geometry.is_empty:
        skipped += 1
        continue

    # ── Extract coordinates ───────────────────────────────────────────────
    # Folium needs coordinates as [(lat, lon), (lat, lon), ...]
    # Shapely stores them as (lon, lat) — must swap
    geom = row.geometry

    if geom.geom_type == 'LineString':
        coords = [(lat, lon) for lon, lat in geom.coords]

    elif geom.geom_type == 'MultiLineString':
        # MultiLineString = multiple disconnected line pieces
        # Flatten all pieces into one coordinate list
        coords = []
        for line in geom.geoms:
            coords.extend([(lat, lon) for lon, lat in line.coords])

    else:
        skipped += 1
        continue

    if len(coords) < 2:
        skipped += 1
        continue

    # ── Get style values ──────────────────────────────────────────────────
    risk_score    = row.get('risk_score', 0)
    risk_category = row.get('risk_category', 'No Data')
    road_name     = row.get('name', 'Unnamed Road')
    highway_type  = row.get('highway', 'unclassified')
    length_m      = row.get('length', 0)
    total_comp    = int(row.get('total_complaints', 0))
    pothole_c     = int(row.get('pothole_count', 0))
    high_sev      = int(row.get('high_severity_count', 0))
    monsoon_c     = int(row.get('monsoon_complaints', 0))

    # Clean up list-type values that OSMnx sometimes produces
    if isinstance(road_name,   list): road_name   = road_name[0]
    if isinstance(highway_type,list): highway_type= highway_type[0]

    road_name    = str(road_name)    if road_name    else 'Unnamed Road'
    highway_type = str(highway_type) if highway_type else 'unclassified'

    # ── Risk score display ────────────────────────────────────────────────
    score_display = f"{risk_score:.1f}" if risk_score > 0 else "No data"

    # ── Colour-coded risk badge for tooltip ───────────────────────────────
    badge_colors = {
        'High':    ('background:#FADBD8;color:#C0392B', '🔴'),
        'Medium':  ('background:#FDEBD0;color:#D35400', '🟡'),
        'Low':     ('background:#D5F5E3;color:#1E8449', '🟢'),
        'No Data': ('background:#EAECEE;color:#7F8C8D', '⚪'),
    }
    badge_style, badge_icon = badge_colors.get(
        risk_category, ('background:#EAECEE;color:#7F8C8D', '⚪')
    )

    # ── Build tooltip HTML ────────────────────────────────────────────────
    tooltip_html = f"""
    <div style="font-family:Arial; min-width:200px; max-width:260px;">

      <div style="font-size:14px; font-weight:bold; color:#1B2631;
                  border-bottom:2px solid #1A5276; padding-bottom:6px;
                  margin-bottom:8px;">
        {road_name}
      </div>

      <div style="display:inline-block; padding:3px 10px;
                  border-radius:12px; font-size:12px;
                  font-weight:bold; margin-bottom:8px; {badge_style}">
        {badge_icon} {risk_category} Risk
      </div>

      <table style="width:100%; font-size:12px; border-collapse:collapse;">
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">Risk Score</td>
          <td style="font-weight:bold; text-align:right;">
            {score_display} / 10
          </td>
        </tr>
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">Road Type</td>
          <td style="text-align:right;">{highway_type}</td>
        </tr>
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">Length</td>
          <td style="text-align:right;">{length_m:.0f} m</td>
        </tr>
        <tr style="border-top:1px solid #EAECEE; margin-top:4px;">
          <td style="color:#7F8C8D; padding:4px 0 2px;">Total Complaints</td>
          <td style="font-weight:bold; text-align:right;">{total_comp}</td>
        </tr>
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">Potholes</td>
          <td style="text-align:right;">{pothole_c}</td>
        </tr>
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">High Severity</td>
          <td style="text-align:right;">{high_sev}</td>
        </tr>
        <tr>
          <td style="color:#7F8C8D; padding:2px 0;">Monsoon Reports</td>
          <td style="text-align:right;">{monsoon_c}</td>
        </tr>
      </table>
    </div>
    """

    # ── Add polyline to layer ─────────────────────────────────────────────
    folium.PolyLine(
        locations=coords,
        color=get_road_color(risk_score),
        weight=get_road_weight(highway_type),
        opacity=get_road_opacity(risk_score),
        tooltip=folium.Tooltip(
            tooltip_html,
            sticky=True   # Tooltip follows mouse, doesn't snap to segment
        )
    ).add_to(road_layer)

    added += 1

road_layer.add_to(m)

print(f"Road layer complete!")
print(f"  Added:   {added:,} road segments")
print(f"  Skipped: {skipped} (null or empty geometries)")
````

````python
# ── Helper functions for visualization ───────────────────────────────────

def get_severity_radius(sev):
    if str(sev).lower() == 'high':
        return 8
    elif str(sev).lower() == 'medium':
        return 6
    else:
        return 4

def get_complaint_color(ctype):
    c = str(ctype).lower()
    
    if 'pothole' in c:
        return '#E74C3C'   # red
    elif 'water' in c or 'waterlog' in c:
        return '#3498DB'   # blue
    elif 'damage' in c:
        return '#F39C12'   # orange
    else:
        return '#7F8C8D'   # grey
````

````python
# ── Layer 2: Individual complaint points ───────────────────────────────────

complaint_layer = folium.FeatureGroup(
    name="📍 Complaint Points",
    show=False   # Hidden by default — user can toggle on
)

for idx, row in complaints.iterrows():

    if row.geometry is None:
        continue

    lat  = row.geometry.y
    lon  = row.geometry.x
    ctype= row.get('complaint_type', 'Unknown')
    sev  = row.get('severity', 'Unknown')
    month= row.get('month', 0)
    mon  = row.get('is_monsoon', 0)

    month_names = {
        1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
        7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'
    }
    month_label = month_names.get(int(month), str(month))
    monsoon_label = '🌧️ Monsoon month' if mon == 1 else '☀️ Dry season'

    complaint_tooltip = f"""
    <div style="font-family:Arial; font-size:12px; min-width:160px;">
      <b style="color:#1B2631;">{ctype}</b><br>
      <span style="color:#7F8C8D;">Severity:</span>
      <b style="color:{'#E74C3C' if sev=='High' else '#F39C12' if sev=='Medium' else '#27AE60'};">
        {sev}
      </b><br>
      <span style="color:#7F8C8D;">Month:</span> {month_label}
      &nbsp; {monsoon_label}
    </div>
    """

    folium.CircleMarker(
        location=[lat, lon],
        radius=get_severity_radius(sev),
        color=get_complaint_color(ctype),
        fill=True,
        fill_color=get_complaint_color(ctype),
        fill_opacity=0.75,
        weight=0.5,
        tooltip=folium.Tooltip(complaint_tooltip, sticky=False)
    ).add_to(complaint_layer)

complaint_layer.add_to(m)
print(f"Complaint layer added: {len(complaints):,} points")
````

````python
# ── Layer 3: High risk road highlights ────────────────────────────────────

highlight_layer = folium.FeatureGroup(
    name="🚨 High Risk Roads Only",
    show=False
)

high_risk_roads = roads[roads['risk_category'] == 'High'].copy()

for idx, row in high_risk_roads.iterrows():

    if row.geometry is None or row.geometry.is_empty:
        continue

    geom = row.geometry
    if geom.geom_type == 'LineString':
        coords = [(lat, lon) for lon, lat in geom.coords]
    elif geom.geom_type == 'MultiLineString':
        coords = []
        for line in geom.geoms:
            coords.extend([(lat, lon) for lon, lat in line.coords])
    else:
        continue

    road_name = row.get('name', 'Unnamed Road')
    if isinstance(road_name, list): road_name = road_name[0]
    road_name = str(road_name) if road_name else 'Unnamed Road'

    risk_score    = row.get('risk_score', 0)
    total_comp    = int(row.get('total_complaints', 0))

    # Pulse effect — thicker white outline + red fill
    # Draw white line first (slightly wider) then red on top
    folium.PolyLine(
        locations=coords,
        color='#FFFFFF',
        weight=8,
        opacity=0.8
    ).add_to(highlight_layer)

    folium.PolyLine(
        locations=coords,
        color='#C0392B',
        weight=5,
        opacity=0.95,
        tooltip=folium.Tooltip(
            f"<b style='color:#C0392B;'>⚠️ HIGH RISK</b><br>"
            f"<b>{road_name}</b><br>"
            f"Score: {risk_score:.1f}/10<br>"
            f"Complaints: {total_comp}",
            sticky=True
        )
    ).add_to(highlight_layer)

highlight_layer.add_to(m)
print(f"High risk highlight layer: {len(high_risk_roads)} roads")
````

````python
# ── Legend ─────────────────────────────────────────────────────────────────

cat_counts = roads['risk_category'].value_counts()

def safe_count(cat):
    return cat_counts.get(cat, 0)

legend_html = f"""
<div style="
    position: fixed;
    bottom: 40px;
    left: 40px;
    z-index: 1000;
    background: white;
    padding: 16px 20px;
    border-radius: 10px;
    border: 1px solid #D5D8DC;
    font-family: Arial, sans-serif;
    font-size: 13px;
    box-shadow: 2px 2px 8px rgba(0,0,0,0.15);
    min-width: 200px;
">
  <div style="font-size:15px; font-weight:bold; color:#1A5276;
              border-bottom:2px solid #1A5276; padding-bottom:8px;
              margin-bottom:12px;">
    Road Risk Score
  </div>

  <div style="margin-bottom:8px; display:flex;
              align-items:center; gap:10px;">
    <div style="width:30px; height:5px; background:#E74C3C;
                border-radius:3px;"></div>
    <span>High Risk (7–10)
      <span style="color:#7F8C8D; font-size:11px;">
        — {safe_count('High')} roads
      </span>
    </span>
  </div>

  <div style="margin-bottom:8px; display:flex;
              align-items:center; gap:10px;">
    <div style="width:30px; height:5px; background:#F39C12;
                border-radius:3px;"></div>
    <span>Medium Risk (4–7)
      <span style="color:#7F8C8D; font-size:11px;">
        — {safe_count('Medium')} roads
      </span>
    </span>
  </div>

  <div style="margin-bottom:8px; display:flex;
              align-items:center; gap:10px;">
    <div style="width:30px; height:5px; background:#27AE60;
                border-radius:3px;"></div>
    <span>Low Risk (1–4)
      <span style="color:#7F8C8D; font-size:11px;">
        — {safe_count('Low')} roads
      </span>
    </span>
  </div>

  <div style="margin-bottom:12px; display:flex;
              align-items:center; gap:10px;">
    <div style="width:30px; height:5px; background:#BDC3C7;
                border-radius:3px;"></div>
    <span>No Data
      <span style="color:#7F8C8D; font-size:11px;">
        — {safe_count('No Data')} roads
      </span>
    </span>
  </div>

  <div style="border-top:1px solid #EAECEE; padding-top:10px;
              font-size:11px; color:#7F8C8D; line-height:1.5;">
    <b>Formula:</b><br>
    0.5 × pothole density<br>
    + 0.3 × traffic volume<br>
    + 0.2 × road vulnerability<br>
    <br>
    <b>Data:</b> OpenStreetMap + BMC<br>
    <b>Area:</b> Bandra West, Mumbai
  </div>
</div>
"""

m.get_root().html.add_child(folium.Element(legend_html))
print("Legend added!")
````

````python
# ── Key findings panel ─────────────────────────────────────────────────────

high_pct  = safe_count('High')  / len(roads) * 100
med_pct   = safe_count('Medium')/ len(roads) * 100

# Worst road name
worst_road = roads.nlargest(1, 'risk_score').iloc[0]
worst_name = worst_road.get('name', 'Unnamed')
if isinstance(worst_name, list): worst_name = worst_name[0]
worst_name = str(worst_name) if worst_name else 'Unnamed Road'
worst_score= worst_road['risk_score']

# Monsoon stat
monsoon_total   = complaints['is_monsoon'].sum()
monsoon_pct     = monsoon_total / len(complaints) * 100

# Residential roads
res_mask = roads['highway'].astype(str).str.contains('residential', na=False)
res_roads = roads[res_mask]
res_high  = res_roads[res_roads['risk_category'] == 'High']
res_high_pct = len(res_high) / max(len(res_roads), 1) * 100

findings_html = f"""
<div style="
    position: fixed;
    top: 20px;
    right: 50px;
    z-index: 1000;
    background: white;
    padding: 16px 18px;
    border-radius: 10px;
    border: 1px solid #D5D8DC;
    font-family: Arial, sans-serif;
    font-size: 12px;
    box-shadow: 2px 2px 8px rgba(0,0,0,0.15);
    max-width: 280px;
">
  <div style="font-size:14px; font-weight:bold; color:#1A5276;
              border-bottom:2px solid #1A5276; padding-bottom:6px;
              margin-bottom:10px;">
    📊 Key Findings
  </div>

  <div style="background:#FDEDEC; border-left:3px solid #E74C3C;
              padding:8px 10px; border-radius:4px; margin-bottom:8px;">
    <b style="color:#C0392B;">🔴 {high_pct:.0f}% of roads</b>
    <span style="color:#2C3E50;"> are High Risk
    ({safe_count('High')} segments)</span>
  </div>

  <div style="background:#FEF9E7; border-left:3px solid #F39C12;
              padding:8px 10px; border-radius:4px; margin-bottom:8px;">
    <b style="color:#D35400;">⚠️ Worst road:</b>
    <span style="color:#2C3E50;"> {worst_name[:30]}
    (score: {worst_score:.1f})</span>
  </div>

  <div style="background:#EBF5FB; border-left:3px solid #3498DB;
              padding:8px 10px; border-radius:4px; margin-bottom:8px;">
    <b style="color:#1A5276;">🌧️ {monsoon_pct:.0f}% of complaints</b>
    <span style="color:#2C3E50;"> occur in monsoon months (Jun–Sep)</span>
  </div>

  <div style="background:#EAFAF1; border-left:3px solid #27AE60;
              padding:8px 10px; border-radius:4px; margin-bottom:8px;">
    <b style="color:#1E8449;">🏘️ Residential roads:</b>
    <span style="color:#2C3E50;"> {res_high_pct:.0f}% are High Risk
    despite lower traffic</span>
  </div>

  <div style="border-top:1px solid #EAECEE; padding-top:8px;
              color:#7F8C8D; font-size:11px;">
    Hover over any road for details.<br>
    Use layers panel (top right) to toggle views.
  </div>
</div>
"""

m.get_root().html.add_child(folium.Element(findings_html))
print("Key findings panel added!")
````

````python
# ── Layer control — MUST be added last ────────────────────────────────────
folium.LayerControl(
    position='topright',
    collapsed=False   # Show expanded by default so user sees the options
).add_to(m)

# ── Save the map ───────────────────────────────────────────────────────────
os.makedirs("../outputs", exist_ok=True)
output_path = "../outputs/mumbai_road_risk_map.html"
m.save(output_path)

file_size = os.path.getsize(output_path) / 1024
print(f"Map saved!")
print(f"  Path:      {output_path}")
print(f"  File size: {file_size:.0f} KB")
print(f"\nOpen this file in Chrome or Firefox to view your map.")
````

Goal: Build analytics charts, generate automated insights, publish to GitHub

````python
# ── Phase 6 Imports ───────────────────────────────────────────────────────
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
import os
warnings.filterwarnings('ignore')

# Set a clean matplotlib style for all charts
plt.rcParams.update({
    'font.family':       'Arial',
    'font.size':         11,
    'axes.spines.top':   False,
    'axes.spines.right': False,
    'axes.grid':         True,
    'grid.alpha':        0.3,
    'grid.linestyle':    '--',
    'figure.facecolor':  'white',
    'axes.facecolor':    'white',
})

# ── Reload all data ────────────────────────────────────────────────────────
roads      = gpd.read_file("../data/roads_scored.gpkg")
complaints = gpd.read_file("../data/complaints_bandra.gpkg")

# Convert complaints GeoDataFrame back to plain DataFrame for charting
complaints_df = pd.DataFrame(complaints.drop(columns='geometry'))

print("All data reloaded!")
print(f"  Roads:      {len(roads):,} segments")
print(f"  Complaints: {len(complaints_df):,} records")
print(f"\nRisk category breakdown:")
print(roads['risk_category'].value_counts())
````

````python
# ── Chart 1: Risk category donut chart ────────────────────────────────────

fig, ax = plt.subplots(figsize=(8, 6))

cat_order  = ['High', 'Medium', 'Low', 'No Data']
cat_colors = {
    'High':    '#E74C3C',
    'Medium':  '#F39C12',
    'Low':     '#27AE60',
    'No Data': '#BDC3C7',
}

cat_counts = roads['risk_category'].value_counts()
cats       = [c for c in cat_order if c in cat_counts.index]
counts     = [cat_counts[c] for c in cats]
colors     = [cat_colors[c] for c in cats]
total      = sum(counts)

wedges, texts, autotexts = ax.pie(
    counts,
    labels=None,
    colors=colors,
    autopct=lambda pct: f'{pct:.1f}%\n({int(round(pct*total/100))})',
    pctdistance=0.75,
    startangle=90,
    wedgeprops=dict(width=0.55, edgecolor='white', linewidth=2),
    textprops=dict(fontsize=10)
)

# Style the percentage labels
for autotext, cat in zip(autotexts, cats):
    autotext.set_color('white' if cat in ['High', 'Medium', 'Low'] else '#555')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(10)

# Centre label
ax.text(0, 0.05, str(total), ha='center', va='center',
        fontsize=28, fontweight='bold', color='#1B2631')
ax.text(0, -0.18, 'road segments', ha='center', va='center',
        fontsize=11, color='#7F8C8D')

# Custom legend
legend_elements = [
    mpatches.Patch(color=cat_colors[c],
                   label=f"{c} Risk — {cat_counts.get(c,0)} roads")
    for c in cats
]
ax.legend(handles=legend_elements, loc='lower center',
          bbox_to_anchor=(0.5, -0.08), ncol=2,
          fontsize=10, frameon=False)

ax.set_title('Road Risk Distribution — Bandra West, Mumbai',
             fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.savefig("../outputs/chart1_risk_distribution.png",
            dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("Chart 1 saved!")
````

````python
# ── Chart 2: Top 10 highest risk road segments ────────────────────────────

fig, ax = plt.subplots(figsize=(11, 6))

# Get top 10 by risk score
top10 = roads.nlargest(10, 'risk_score').copy()

# Clean road names
def clean_name(name_val):
    if isinstance(name_val, list): name_val = name_val[0]
    if pd.isna(name_val) or str(name_val) == 'nan': return 'Unnamed Road'
    return str(name_val)[:35]

top10['clean_name']    = top10['name'].apply(clean_name)
top10['highway_clean'] = top10['highway'].apply(
    lambda x: str(x[0] if isinstance(x, list) else x)
)

# Assign bar colours by risk score
bar_colors = [
    '#C0392B' if s >= 8 else '#E74C3C' if s >= 7 else '#F39C12'
    for s in top10['risk_score']
]

bars = ax.barh(
    range(len(top10)),
    top10['risk_score'],
    color=bar_colors,
    edgecolor='white',
    linewidth=0.5,
    height=0.65
)

# Labels inside/outside bars
for i, (bar, row) in enumerate(zip(bars, top10.itertuples())):
    width = bar.get_width()

    # Road type badge on the left
    ax.text(-0.15, i, f"[{row.highway_clean}]",
            va='center', ha='right', fontsize=8,
            color='#7F8C8D', style='italic')

    # Score label at end of bar
    ax.text(width + 0.05, i, f"{width:.1f}",
            va='center', ha='left', fontsize=10,
            fontweight='bold', color='#1B2631')

    # Total complaints annotation
    ax.text(0.15, i, f"{int(row.total_complaints)} complaints",
            va='center', ha='left', fontsize=8.5,
            color='white', fontweight='bold')

ax.set_yticks(range(len(top10)))
ax.set_yticklabels(top10['clean_name'], fontsize=10)
ax.set_xlabel('Risk Score (0–10)', fontsize=11)
ax.set_title('Top 10 Highest Risk Road Segments — Bandra West',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlim(0, 11.5)
ax.axvline(x=7, color='#E74C3C', linestyle='--',
           linewidth=1.2, alpha=0.6, label='High risk threshold (7.0)')
ax.axvline(x=4, color='#F39C12', linestyle='--',
           linewidth=1.2, alpha=0.6, label='Medium risk threshold (4.0)')
ax.legend(fontsize=9, loc='lower right')
ax.invert_yaxis()   # Worst road at top

plt.tight_layout()
plt.savefig("../outputs/chart2_top10_roads.png",
            dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("Chart 2 saved!")
````

````python
# ── Chart 3: Complaint burden vs network share by road type ───────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Clean highway column
roads['highway_clean'] = roads['highway'].apply(
    lambda x: str(x[0] if isinstance(x, list) else x).strip()
)

# Road type stats
road_type_stats = roads.groupby('highway_clean').agg(
    segment_count     = ('risk_score',       'count'),
    total_complaints  = ('total_complaints', 'sum'),
    avg_risk          = ('risk_score',       'mean'),
    high_risk_count   = ('risk_category',
                         lambda x: (x == 'High').sum())
).reset_index()

road_type_stats = road_type_stats[
    road_type_stats['segment_count'] >= 5
].sort_values('total_complaints', ascending=False).head(7)

total_segments   = road_type_stats['segment_count'].sum()
total_complaints = road_type_stats['total_complaints'].sum()

road_type_stats['network_share_pct']   = (
    road_type_stats['segment_count']   / total_segments   * 100
)
road_type_stats['complaint_share_pct'] = (
    road_type_stats['total_complaints']/ total_complaints * 100
)

# ── Left: Network share vs complaint share ────────────────────────────────
ax1 = axes[0]
x    = np.arange(len(road_type_stats))
w    = 0.38

bars1 = ax1.bar(x - w/2, road_type_stats['network_share_pct'],
                w, label='% of road network', color='#3498DB', alpha=0.85)
bars2 = ax1.bar(x + w/2, road_type_stats['complaint_share_pct'],
                w, label='% of complaints',   color='#E74C3C', alpha=0.85)

ax1.set_xticks(x)
ax1.set_xticklabels(road_type_stats['highway_clean'],
                    rotation=30, ha='right', fontsize=9)
ax1.set_ylabel('Percentage (%)')
ax1.set_title('Network Share vs Complaint Burden\nby Road Type',
              fontsize=12, fontweight='bold')
ax1.legend(fontsize=9)

# Annotate the gap — where complaint share >> network share
for i, row in enumerate(road_type_stats.itertuples()):
    gap = row.complaint_share_pct - row.network_share_pct
    if gap > 3:
        ax1.annotate(
            f'+{gap:.0f}%',
            xy=(i + w/2, row.complaint_share_pct),
            xytext=(0, 5), textcoords='offset points',
            ha='center', fontsize=8,
            color='#C0392B', fontweight='bold'
        )

# ── Right: Average risk score by road type ────────────────────────────────
ax2 = axes[1]
sorted_stats = road_type_stats.sort_values('avg_risk', ascending=True)
bar_colors2  = [
    '#E74C3C' if r >= 7 else '#F39C12' if r >= 4 else '#27AE60'
    for r in sorted_stats['avg_risk']
]

bars3 = ax2.barh(
    sorted_stats['highway_clean'],
    sorted_stats['avg_risk'],
    color=bar_colors2, edgecolor='white', height=0.6
)

for bar, val in zip(bars3, sorted_stats['avg_risk']):
    ax2.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
             f'{val:.1f}', va='center', fontsize=9, fontweight='bold')

ax2.set_xlabel('Average Risk Score (0–10)')
ax2.set_title('Average Risk Score\nby Road Type',
              fontsize=12, fontweight='bold')
ax2.axvline(x=7, color='#E74C3C', linestyle='--', alpha=0.5, linewidth=1)
ax2.axvline(x=4, color='#F39C12', linestyle='--', alpha=0.5, linewidth=1)
ax2.set_xlim(0, 11)

plt.suptitle('Road Type Analysis — Complaint Burden & Risk',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig("../outputs/chart3_road_type_analysis.png",
            dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("Chart 3 saved!")
````

````python
# ── Chart 4: Monthly complaint trend with monsoon annotation ──────────────

fig, ax = plt.subplots(figsize=(12, 5))

month_names = {
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',  5:'May',  6:'Jun',
    7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
}

monthly = complaints_df.groupby('month').agg(
    total          = ('complaint_type',  'count'),
    potholes       = ('complaint_type',
                      lambda x: (x=='Pothole').sum()),
    high_severity  = ('severity',
                      lambda x: (x=='High').sum()),
).reset_index()

monthly['month_name'] = monthly['month'].map(month_names)
monthly_bar_colors = [
    '#E74C3C' if m in [6,7,8,9] else '#5DADE2'
    for m in monthly['month']
]

# Bar chart — total per month
bars = ax.bar(monthly['month_name'], monthly['total'],
              color=monthly_bar_colors,
              edgecolor='white', linewidth=0.5,
              alpha=0.9, zorder=3, label='Total complaints')

# Line overlay — potholes specifically
ax2_twin = ax.twinx()
ax2_twin.plot(monthly['month_name'], monthly['potholes'],
              color='#884EA0', linewidth=2.5,
              marker='o', markersize=6,
              zorder=4, label='Pothole reports')
ax2_twin.plot(monthly['month_name'], monthly['high_severity'],
              color='#D35400', linewidth=2,
              marker='s', markersize=5,
              linestyle='--', zorder=4, label='High severity')
ax2_twin.set_ylabel('Pothole / High Severity Count', color='#884EA0')
ax2_twin.tick_params(axis='y', labelcolor='#884EA0')

# Monsoon shading
ax.axvspan(4.5, 8.5, alpha=0.07, color='#3498DB', zorder=1)
ax.text(6.5, monthly['total'].max() * 0.92,
        '🌧️ Monsoon Season\n(Jun – Sep)',
        ha='center', fontsize=10, color='#1A5276',
        fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.4', facecolor='#EBF5FB',
                  edgecolor='#3498DB', alpha=0.8))

# Value labels on bars
for bar in bars:
    h = bar.get_height()
    if h > 0:
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.5,
                str(int(h)), ha='center', va='bottom',
                fontsize=8.5, color='#1B2631', fontweight='bold')

ax.set_xlabel('')
ax.set_ylabel('Total Complaints', fontsize=11)
ax.set_title('Monthly Complaint Pattern — Seasonal Risk Variation',
             fontsize=14, fontweight='bold', pad=15)

# Combined legend from both axes
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2,
          loc='upper left', fontsize=9, framealpha=0.9)

# Colour legend for bars
blue_patch = mpatches.Patch(color='#5DADE2', alpha=0.9, label='Dry season')
red_patch  = mpatches.Patch(color='#E74C3C', alpha=0.9, label='Monsoon')
ax.legend(handles=[blue_patch, red_patch] + lines2,
          labels=['Dry season months', 'Monsoon months'] + labels2,
          loc='upper left', fontsize=9, framealpha=0.9)

plt.tight_layout()
plt.savefig("../outputs/chart4_monthly_trend.png",
            dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("Chart 4 saved!")
````

````python
# ── Chart 5: What drives risk — factor contribution heatmap ───────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ── Left: Correlation between factors and final risk score ────────────────
ax1 = axes[0]

factor_cols = ['f1_norm', 'f2_norm', 'f3_norm', 'risk_score']
col_labels  = {
    'f1_norm':    'Pothole\nDensity (F1)',
    'f2_norm':    'Traffic\nVolume (F2)',
    'f3_norm':    'Road\nVulnerability (F3)',
    'risk_score': 'Final\nRisk Score'
}

available = [c for c in factor_cols if c in roads.columns]
corr_data = roads[available].corr()
corr_data.index   = [col_labels.get(c, c) for c in corr_data.index]
corr_data.columns = [col_labels.get(c, c) for c in corr_data.columns]

im = ax1.imshow(corr_data.values, cmap='RdYlGn',
                vmin=-1, vmax=1, aspect='auto')

ax1.set_xticks(range(len(corr_data.columns)))
ax1.set_yticks(range(len(corr_data.index)))
ax1.set_xticklabels(corr_data.columns, fontsize=9)
ax1.set_yticklabels(corr_data.index, fontsize=9)

for i in range(len(corr_data)):
    for j in range(len(corr_data.columns)):
        val = corr_data.values[i, j]
        ax1.text(j, i, f'{val:.2f}',
                 ha='center', va='center',
                 fontsize=10, fontweight='bold',
                 color='white' if abs(val) > 0.5 else '#1B2631')

plt.colorbar(im, ax=ax1, shrink=0.8, label='Correlation coefficient')
ax1.set_title('Factor Correlation Matrix',
              fontsize=12, fontweight='bold')

# ── Right: Average factor values by risk category ─────────────────────────
ax2 = axes[1]

cat_factor = roads.groupby('risk_category')[
    [c for c in ['f1_norm','f2_norm','f3_norm'] if c in roads.columns]
].mean()

cat_factor = cat_factor.reindex(
    [c for c in ['Low','Medium','High'] if c in cat_factor.index]
)
cat_factor.index = [f"{c} Risk" for c in cat_factor.index]
if 'f1_norm' in cat_factor.columns: cat_factor = cat_factor.rename(
    columns={
        'f1_norm': 'Pothole Density\n(50%)',
        'f2_norm': 'Traffic Volume\n(30%)',
        'f3_norm': 'Road Vulnerability\n(20%)'
    }
)

x     = np.arange(len(cat_factor))
w     = 0.26
cols  = cat_factor.columns.tolist()
fc    = ['#E74C3C', '#3498DB', '#F39C12']

for i, (col, color) in enumerate(zip(cols, fc)):
    offset = (i - 1) * w
    bars   = ax2.bar(x + offset, cat_factor[col],
                     w, label=col,
                     color=color, alpha=0.85,
                     edgecolor='white')
    for bar in bars:
        h = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2, h + 0.005,
                 f'{h:.2f}', ha='center', va='bottom',
                 fontsize=8)

ax2.set_xticks(x)
ax2.set_xticklabels(cat_factor.index, fontsize=10)
ax2.set_ylabel('Average Normalised Factor Value (0–1)')
ax2.set_title('Factor Contributions\nby Risk Category',
              fontsize=12, fontweight='bold')
ax2.legend(fontsize=8, loc='upper left')
ax2.set_ylim(0, 1.1)

plt.suptitle('Risk Score Analysis — Factor Deep Dive',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig("../outputs/chart5_factor_analysis.png",
            dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print("Chart 5 saved!")
````

````python
# ── Automated Insights Engine ──────────────────────────────────────────────

insights = []

# ── Helper: safe percentage ────────────────────────────────────────────────
def pct(numerator, denominator):
    if denominator == 0: return 0
    return round(numerator / denominator * 100, 1)

total_roads      = len(roads)
total_complaints = int(roads['total_complaints'].sum())
cat_counts       = roads['risk_category'].value_counts()

# ── Insight 1: Overall risk landscape ────────────────────────────────────
high_n   = cat_counts.get('High',    0)
medium_n = cat_counts.get('Medium',  0)
low_n    = cat_counts.get('Low',     0)
no_data_n= cat_counts.get('No Data', 0)

insights.append({
    'id':       1,
    'category': 'Overall Risk Landscape',
    'icon':     '🗺️',
    'finding':  (
        f"{pct(high_n, total_roads)}% of road segments in Bandra West "
        f"({high_n} of {total_roads}) are classified as High Risk, "
        f"while {pct(low_n, total_roads)}% are Low Risk. "
        f"This suggests a road network under significant stress "
        f"in approximately one third of its coverage area."
    ),
    'implication': (
        "Immediate maintenance intervention is warranted on "
        f"at least {high_n} road segments."
    )
})

# ── Insight 2: Residential road burden ────────────────────────────────────
roads['highway_str'] = roads['highway'].apply(
    lambda x: str(x[0] if isinstance(x, list) else x)
)

res_roads  = roads[roads['highway_str'].str.contains('residential', na=False)]
res_n      = len(res_roads)
res_comp   = int(res_roads['total_complaints'].sum())
res_high   = int((res_roads['risk_category'] == 'High').sum())

insights.append({
    'id':       2,
    'category': 'Residential Road Burden',
    'icon':     '🏘️',
    'finding':  (
        f"Residential roads make up {pct(res_n, total_roads)}% of the road "
        f"network but account for {pct(res_comp, total_complaints)}% of all "
        f"complaints. {pct(res_high, res_n)}% of residential segments "
        f"are High Risk — the highest proportion of any road category."
    ),
    'implication': (
        "Municipal maintenance budgets disproportionately benefit primary "
        "roads. Residential lanes are systematically underserved."
    )
})

# ── Insight 3: Monsoon seasonality ────────────────────────────────────────
monsoon_total    = int(complaints_df['is_monsoon'].sum())
non_monsoon      = len(complaints_df) - monsoon_total
monsoon_pct_val  = pct(monsoon_total, len(complaints_df))

peak_month_num   = complaints_df['month'].value_counts().index[0]
month_names_dict = {
    1:'January', 2:'February', 3:'March', 4:'April',
    5:'May', 6:'June', 7:'July', 8:'August',
    9:'September', 10:'October', 11:'November', 12:'December'
}
peak_month_name  = month_names_dict.get(int(peak_month_num), str(peak_month_num))

insights.append({
    'id':       3,
    'category': 'Monsoon Seasonality',
    'icon':     '🌧️',
    'finding':  (
        f"{monsoon_pct_val}% of all complaints occur during monsoon months "
        f"(June–September), with {peak_month_name} recording the highest "
        f"single-month complaint volume. Non-monsoon months average "
        f"{non_monsoon//8:.0f} complaints per month vs "
        f"{monsoon_total//4:.0f} during monsoon — "
        f"a {round(monsoon_total/4 / max(non_monsoon/8,1)):.0f}× increase."
    ),
    'implication': (
        "Road deterioration is predictable and seasonal. "
        "Pre-monsoon maintenance (March–May) would reduce peak-season "
        "complaint volumes significantly."
    )
})

# ── Insight 4: Severity concentration ─────────────────────────────────────
high_sev_total = int(complaints_df[complaints_df['severity']=='High'].shape[0])
high_sev_pct   = pct(high_sev_total, len(complaints_df))

top_10_pct_roads = roads.nlargest(int(total_roads * 0.10), 'total_complaints')
top_10_complaints= int(top_10_pct_roads['total_complaints'].sum())

insights.append({
    'id':       4,
    'category': 'Complaint Severity & Concentration',
    'icon':     '⚠️',
    'finding':  (
        f"{high_sev_pct}% of complaints are classified High Severity "
        f"({high_sev_total} incidents). Critically, the top 10% of road "
        f"segments by complaint volume account for "
        f"{pct(top_10_complaints, total_complaints)}% of all complaints — "
        f"a classic Pareto distribution pattern."
    ),
    'implication': (
        "Targeting maintenance on the top 10% of roads by complaint count "
        f"would address {pct(top_10_complaints, total_complaints)}% of "
        "all reported issues — maximum impact for minimum resource spend."
    )
})

# ── Insight 5: Primary roads — low volume, high impact ────────────────────
prim_roads     = roads[roads['highway_str'].str.contains('primary', na=False)]
prim_avg_risk  = prim_roads['risk_score'].mean()
prim_avg_comp  = prim_roads['total_complaints'].mean()
prim_high      = int((prim_roads['risk_category']=='High').sum())

insights.append({
    'id':       5,
    'category': 'Primary Road Risk',
    'icon':     '🛣️',
    'finding':  (
        f"Primary roads show an average risk score of {prim_avg_risk:.1f}/10 "
        f"with an average of {prim_avg_comp:.1f} complaints per segment. "
        f"{prim_high} primary road segments are High Risk — "
        f"significant because primary roads carry the highest traffic volumes "
        f"and their degradation affects the most road users."
    ),
    'implication': (
        "Primary road damage has outsized impact. Even moderate "
        "deterioration on a primary road affects thousands of "
        "daily commuters."
    )
})

# ── Insight 6: Risk-aware routing benefit ─────────────────────────────────
avg_risk_all  = roads['risk_score'].mean()
avg_risk_high = roads[roads['risk_category']=='High']['risk_score'].mean()

insights.append({
    'id':       6,
    'category': 'Routing Intelligence Opportunity',
    'icon':     '🗺️',
    'finding':  (
        f"The average network risk score is {avg_risk_all:.1f}/10. "
        f"High-risk segments average {avg_risk_high:.1f}/10. "
        f"Risk-aware routing that avoids the top {pct(high_n, total_roads):.0f}% "
        f"of road segments could meaningfully reduce vehicle wear, "
        f"improve passenger comfort, and reduce accident probability "
        f"on affected routes."
    ),
    'implication': (
        "This is the core value proposition of HERE's routing products — "
        "road quality as a routing parameter alongside distance and speed."
    )
})

# ── Print all insights ────────────────────────────────────────────────────
print("=" * 65)
print("  AUTOMATED INSIGHTS — BANDRA WEST ROAD RISK ANALYSIS")
print("=" * 65)

for ins in insights:
    print(f"\n{ins['icon']} [{ins['id']}] {ins['category'].upper()}")
    print(f"   Finding:     {ins['finding']}")
    print(f"   Implication: {ins['implication']}")

print(f"\n{'='*65}")
print(f"  {len(insights)} insights generated automatically from data")
````

````python
road_layer = folium.FeatureGroup(name="Road Risk Score", show=True)

added = 0
for idx, row in roads.iterrows():
    geom = row.geometry
    if geom is None or geom.is_empty:
        continue

    try:
        if geom.geom_type == 'LineString':
            coords = [(y, x) for x, y in geom.coords]  # x=lon, y=lat
        elif geom.geom_type == 'MultiLineString':
            coords = [(y, x) for line in geom.geoms for x, y in line.coords]
        else:
            continue

        if len(coords) < 2:
            continue

        folium.PolyLine(
            locations=coords,
            color=get_road_color(row.get('risk_score', 0)),
            weight=3,
            opacity=0.9
        ).add_to(road_layer)
        added += 1

    except Exception as e:
        continue

road_layer.add_to(m)
print(f"Added {added} road lines")
````

````python
def get_road_color(risk_score):
    if pd.isna(risk_score) or risk_score == 0:
        return '#808B96'   # darker grey — visible on light basemap
    elif risk_score >= 7.0:
        return '#E74C3C'   # red — high risk
    elif risk_score >= 4.0:
        return '#E67E00'   # deeper orange — more visible than #F39C12
    else:
        return '#1A8C4E'   # darker green — actually visible on grey basemap

def get_road_weight(highway_val):
    if isinstance(highway_val, list):
        highway_val = highway_val[0]
    highway_val = str(highway_val).lower()
    weight_map = {
        'motorway':    7,
        'trunk':       6,
        'primary':     5,
        'secondary':   4,
        'tertiary':    3.5,
        'residential': 3,    # was 2 — too thin
        'unclassified':2.5,
        'service':     2.5,
    }
    return weight_map.get(highway_val, 3)

def get_road_opacity(risk_score):
    if pd.isna(risk_score) or risk_score == 0:
        return 0.5
    elif risk_score >= 7.0:
        return 1.0
    elif risk_score >= 4.0:
        return 1.0
    else:
        return 0.9    # was 0.75 — too faint
````

````python
import geopandas as gpd
import pandas as pd

roads = gpd.read_file("../data/roads_scored.gpkg")

print("=== SHAPE ===")
print(roads.shape)

print("\n=== CRS ===")
print(roads.crs)

print("\n=== GEOMETRY TYPES ===")
print(roads.geometry.geom_type.value_counts())

print("\n=== RISK SCORE STATS ===")
print(roads['risk_score'].describe())

print("\n=== RISK CATEGORIES ===")
print(roads['risk_category'].value_counts())

print("\n=== SAMPLE GEOMETRY (first road) ===")
geom = roads.geometry.iloc[0]
print("Type:", geom.geom_type)
if geom.geom_type == 'LineString':
    print("First 2 coords:", list(geom.coords)[:2])
elif geom.geom_type == 'MultiLineString':
    print("First 2 coords:", list(geom.geoms[0].coords)[:2])

print("\n=== SAMPLE ROW ===")
print(roads[['name','highway','risk_score','risk_category']].head(5).to_string())
````

````python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# ── Step 1: Apply the weighted formula ────────────────────────────────────
W1, W2, W3 = 0.5, 0.3, 0.2

roads['raw_risk'] = (
    W1 * roads['f1_norm'] +
    W2 * roads['f2_norm'] +
    W3 * roads['f3_norm']
)

# ── Step 2: Apply monsoon multiplier ──────────────────────────────────────
monsoon_threshold = roads['monsoon_complaints'].quantile(0.75)
roads['monsoon_multiplier'] = roads['monsoon_complaints'].apply(
    lambda x: 1.20 if x > monsoon_threshold else 1.0
)
roads['raw_risk_boosted'] = roads['raw_risk'] * roads['monsoon_multiplier']

# ── Step 3: Check distribution BEFORE scaling ─────────────────────────────
print("Raw risk distribution:")
print(roads['raw_risk_boosted'].describe())
print(f"\n90th percentile: {roads['raw_risk_boosted'].quantile(0.90):.4f}")
print(f"95th percentile: {roads['raw_risk_boosted'].quantile(0.95):.4f}")
print(f"99th percentile: {roads['raw_risk_boosted'].quantile(0.99):.4f}")
print(f"Max:             {roads['raw_risk_boosted'].max():.4f}")
````

````python
# ── Step 4: Clip outliers at 99th percentile BEFORE scaling ───────────────
# This prevents one extreme road from compressing all others to near-zero
cap_value = roads['raw_risk_boosted'].quantile(0.99)
print(f"Capping outliers at 99th percentile: {cap_value:.4f}")

roads['raw_risk_capped'] = roads['raw_risk_boosted'].clip(upper=cap_value)

# ── Step 5: Now scale to 0–10 ─────────────────────────────────────────────
scaler = MinMaxScaler(feature_range=(0, 10))
roads['risk_score'] = scaler.fit_transform(
    roads[['raw_risk_capped']]
).round(2)

# ── Step 6: Verify the new distribution ───────────────────────────────────
print("\nFinal risk score distribution:")
print(roads['risk_score'].describe().round(2))
print(f"\nScore percentiles:")
for p in [25, 50, 75, 90, 95, 99]:
    print(f"  {p}th percentile: {roads['risk_score'].quantile(p/100):.2f}")

# ── Step 7: Assign categories ─────────────────────────────────────────────
def categorise_risk(score):
    if score >= 7.0:   return 'High'
    elif score >= 4.0: return 'Medium'
    elif score > 0.0:  return 'Low'
    else:              return 'No Data'

roads['risk_category'] = roads['risk_score'].apply(categorise_risk)

print("\nRisk categories after fix:")
print(roads['risk_category'].value_counts())

# ── Step 8: Save ──────────────────────────────────────────────────────────
roads.to_file("../data/roads_scored.gpkg", driver="GPKG")
print("\nSaved! Now re-run Phase 5 to rebuild the map.")
```

**What you should see after this fix:**

Instead of 2181 Low / 21 Medium / 6 High, you should get something like:
```
Low      ~1400
Medium   ~600
High     ~300
No Data  ~106
````
